### DATA CLEANING

TO HANDLE NULL VALUES

```python
df['low'] = df['low'].fillna(df['low'].mean())        
df['close'] = df['close'].fillna(df['close'].median())  
df['adjClose'] = df['adjClose'].fillna(df['adjClose'].bfill())  
df['volume'] = df['volume'].fillna(0)                 
```

## BIVARIATE ANALYSIS 
(USING GROUP_BY)


```python
print(df.columns.tolist())
```

```python
# Average Close Price per Year
df['year'] = df.index.year
avg_close_by_year = df.groupby('year')['close'].mean()
print(avg_close_by_year)
```

```python
#Total Volume per Year
total_volume_by_year = df.groupby('year')['volume'].sum()
print(total_volume_by_year)
```

```python
#Mean High Price by Year
mean_high_by_year = df.groupby('year')['high'].mean()
print(mean_high_by_year)
```

```python
# Standard Deviation of Low Prices per Year (to measure volatility)
low_std_by_year = df.groupby('year')['low'].std()
print(low_std_by_year)
```

### MULTIVARIATE ANALYSIS USING groupby()

```python
yearly_summary = df.groupby('year')[['open', 'close', 'volume']].mean()
print(yearly_summary)
```

```python
#. Max & Min of Adjusted Close and Volume per Year
yearly_stats = df.groupby('year')[['adjClose', 'volume']].agg(['max', 'min'])
print(yearly_stats)
```

```python
#Quarter-wise Close Price Mean 
df['quarter'] = df.index.to_period('Q')
quarterly_close = df.groupby('quarter')['close'].mean()
print(quarterly_close)
```

#### BIVARIATE ANALYSIS using pivot_table()

```python
#Average Close Price by Year
pivot1 = df.pivot_table(values='close', index='year', aggfunc='mean')
print(pivot1)
# Total Volume by Year
pivot2 = df.pivot_table(values='volume', index='year', aggfunc='sum')
print(pivot2)
```

#### MULTIVARIATE ANALYSIS using pivot_table()

```python
#Average Open, Close, and Volume by Year
pivot3 = df.pivot_table(values=['open', 'close', 'volume'], index='year', aggfunc='mean')
print(pivot3)
#Year-wise Max/Min of Adj Close and Adj Volume
pivot5 = df.pivot_table(values=['adjClose', 'adjVolume'], index='year', aggfunc=['max', 'min'])
print(pivot5)
```

#### Volatility via High vs Low Prices

```python
# Compare how close close price is to high vs low
df['closer_to_high'] = abs(df['close'] - df['high']) < abs(df['close'] - df['low'])

# Count how many times it was closer to high vs low
counts = df['closer_to_high'].value_counts()

# Pie chart
labels = ['Closer to High', 'Closer to Low']
colors = ['darkgreen', 'lightgreen']
plt.pie(counts, labels=labels, autopct='%1.1f%%', startangle=140, colors=colors)
plt.title("Volatility Pie: Close Price Near High vs Low")
plt.show()
```

 #### Quarterly Adjusted Close Price Trend (USING AREA GRAPH)

```python

df.index = pd.to_datetime(df.index)
df['quarter'] = df.index.to_period('Q')

# Group and prepare data
quarterly_adj = df.groupby('quarter')['adjClose'].mean()

# Area Plot
quarterly_adj.plot(kind='area', color='indigo', alpha=0.7, figsize=(10, 5))
plt.title("Quarterly Average Adjusted Close Price (Area Plot)")
plt.xlabel("Quarter")
plt.ylabel("Avg Adj Close Price ($)")
plt.grid(True, color='lightgray', linestyle='--', linewidth=0.5)
plt.show()
```

 #### Quarterly Adjusted Close Price Trend (USING COLUMN CHART)

```python
df['quarter'] = df.index.to_period('Q')
quarterly_adj_close = df.groupby('quarter')['adjClose'].mean()
quarterly_adj_close.plot(kind='bar', figsize=(10,5), title='Quarterly Average Adjusted Close Price', color='maroon')
plt.xlabel('Quarter')
plt.ylabel('Avg Adj Close ($)')
plt.xticks(rotation=45)
plt.grid(True, color='lightgray', linestyle='--', linewidth=0.25)
plt.tight_layout()
plt.show()
```

 #### Quarterly Adjusted Close Price Trend (USING DONUT CHART)

```python
# Total volume per year
vol_year = df.groupby('year')['volume'].sum()
# Plot donut chart
colors = ['navy', 'darkred', 'darkgreen', 'slategray', 'indigo', '#222222', 'maroon', 'midnightblue']

plt.figure(figsize=(6, 6))
plt.pie(
    vol_year,
    labels=vol_year.index,
    autopct='%1.1f%%',
    startangle=90,
    wedgeprops={'width': 0.4},
    colors=colors[:len(vol_year) ] # nicer color range
)
plt.title("Donut Chart: Volume Share by Year")
plt.tight_layout()
plt.show()
```
